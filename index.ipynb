One of the trickier tasks in data cleaning is identifying invalid or impossible values. In these cases, you have to apply your domain knowledge rather than any particular computational technique. In this case, we are looking at heights, which are 1-dimensional, positive numbers. In theory we could have a very tiny height close to 0 cm because the hero is microscopic, but it does not make sense that we would have a height below zero.

## 4. Perform Data Aggregation Required to Answer Third Question

Recall that the third question is: *What are the 5 most common superpowers in Marvel Comics vs. DC Comics?*

We'll need to keep using `heroes_and_powers_df` since we require information from both `heroes_df` and `powers_df`.

Your resulting `question_3_df` should contain aggregated data, with columns `Superpower Name`, `Marvel Comics` (containing the count of occurrences in Marvel Comics), and `DC Comics` (containing the count of occurrences in DC Comics). Each row should represent a superpower.

In other words, `question_3_df` should look like this:

Don't worry if the rows or columns are in a different order, all that matters is that you have the right rows and columns with all the data.

***Hint:*** refer to the [documentation for `.groupby`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html) and treat each publisher as a group.

```python
# Your code here (create more cells as needed)
#creating filter for df
marvel_and_dc_filt = (heroes_and_powers_df['Publisher'] == 'Marvel Comics') | \
                                    (heroes_and_powers_df['Publisher'] == 'DC Comics')

question_3_df = heroes_and_powers_df[marvel_and_dc_filt]

#Checking to see if changes applied successfully
question_3_df['Publisher'].value_counts() #success!


#another option using a filter list
comic_publisher = ['Marvel Comics', 'DC Comics']
filt = heroes_and_powers_df['Publisher'].isin(comic_publisher)

question_3_df_alt = heroes_and_powers_df[filt]

question_3_df_alt['Publisher'].value_counts() #also success! 
```

```python
#With publishers filtered, next step is to filter columns to only include 
#DC, Marvel, and power names

#Since powers_and_heroes is transposed we can use the untransposed index of 
#powers_df to create a list of power names
columns_to_keep = list(powers_df.index)

#For question_3_df we also want publisher values, so we will append 'Publisher'
# to list, which functions as a filter
columns_to_keep.append('Publisher')

#Applying col filter to be applied to question_3_df
question_3_df = question_3_df[columns_to_keep]


```

```python
#Columns and data properly filtered. Now need to group data into respective
#publishers and sum of each respective column
question_3_df = question_3_df.groupby('Publisher').sum()

#Transposing columns so as to make publishers column headers
question_3_df = question_3_df.transpose()

question_3_df
```

```python
#Now the index needs to be renamed to powers followd by resetting the index, so
#that powers becomes a column and the df is indexed by integers

question_3_df.index.name = 'Superpower Name'
question_3_df.reset_index(inplace=True)

#Now the publisher col name should be removed
question_3_df.columns.name = None

question_3_df
```
